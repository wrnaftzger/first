```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sns

header = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']

adult_df = pd.read_csv('adult.data', names = header, encoding_errors = 'ignore', delimiter = ' ', on_bad_lines = 'skip')
adult_df.head()

```

```python
adult_df['sex'] = np.where(adult_df["sex"] == "Male,",1,
                    np.where(adult_df["sex"] == "Female,", 0,
                             np.where(adult_df['sex'] == "other,", 2, 'Other')))
adult_df.head()
```

```python
adult_df['income'] = np.where(adult_df['income'] == "<=50K", 0,
                             np.where(adult_df['income'] == '>50K', 1, 'other'))
```

```python
adult_df['race'] = np.where(adult_df['race'] == "White,",0,
                            np.where(adult_df['race'] == 'Black,', 1,
                                     np.where(adult_df['race'] == 'Asian-Pac-Islander,',2,
                                              np.where(adult_df['race'] == 'Amer-Indian-Eskimo,',3,
                                                       np.where(adult_df['race'] == 'other,', 4,
                                                                np.where(adult_df['race'] == "Other,", 4,'Other'))))))
adult_df
                                                       
```

```python
adult_df.dtypes
```

```python
adult_df.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']

X = adult_df.drop(['age', 'fnlwgt', 'education-num', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week'], axis=1)
y = adult_df[['education', 'workclass', 'marital-status', 'occupation', 'relationship', 'native-country']]

pd.set_option('display.max_columns', None)
cat_headers = pd.get_dummies(X)
cat_headers.head()
```

```python
cleaned_df = adult_df.join(cat_headers)
cleaned_df
```

```python
cleaned_df['education-num'] = cleaned_df['education-num'].str.replace(",", "")
cleaned_df['age'] = cleaned_df['age'].str.replace(",", "")
cleaned_df['fnlwgt'] = cleaned_df['fnlwgt'].str.replace(",", "")
cleaned_df['capital-gain'] = cleaned_df['capital-gain'].str.replace(",", "")
cleaned_df['capital-loss'] = cleaned_df['capital-loss'].str.replace(",", "")
cleaned_df['hours-per-week'] = cleaned_df['hours-per-week'].str.replace(",", "")
```

```python

cleaned_df['age'] = pd.to_numeric(cleaned_df['age'])
cleaned_df['fnlwgt'] = pd.to_numeric(cleaned_df['fnlwgt'])
cleaned_df['education-num'] = pd.to_numeric(cleaned_df['education-num'])
cleaned_df['capital-gain'] = pd.to_numeric(cleaned_df['capital-gain'])
cleaned_df['capital-loss'] = pd.to_numeric(cleaned_df['capital-loss'])
cleaned_df['hours-per-week'] = pd.to_numeric(cleaned_df['hours-per-week'])

cleaned_df
```

```python
corr_df = cleaned_df.drop(['workclass', 'education','marital-status', 'occupation','relationship','native-country','income'], axis =1)
corr_df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

corr_df = corr_df[['race','income_1']]
X = np.array(corr_df['race']).reshape(-1,1)
y = np.array(corr_df['income_1']).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=40)
```

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
y_pred = regression_model.predict(X_test)
y_pred
```

```python
accuracy_score = regression_model.score(X_test, y_test)
print(accuracy_score)
```
