```python
def date_time(df, Dt_Customer):
    '''making the date time column numeric'''
    df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format = '%d-%m-%Y')
    return df
df = date_time(df, 'Dt_Customer')
df.dtypes
```

```python
def education_clean(df, Education):
    '''turning the education column into numerical'''
    df['Education'] = np.where(df['Education'] == 'Graduation', 0,
                               np.where(df['Education'] == 'PhD', 1,
                                        np.where(df['Education'] == 'Master', 2,
                                                 np.where(df['Education'] == 'Basic', 3,
                                                          np.where(df['Education'] ==  '2n Cycle', 4, 0)))))
    return df

df = education_clean(df,'Education')
                                
```

```python
df['Education'].value_counts()
```

```python
def marital_clean(df, Marital_Status):
    '''turning the marital column into numerical data'''
    df['Marital_Status'] = np.where(df['Marital_Status'] == 'Single', 0,
                                    np.where(df['Marital_Status'] == 'Married', 1,
                                             np.where(df['Marital_Status'] == 'Together', 2,
                                                      np.where(df['Marital_Status'] == 'Divorced', 3,
                                                               np.where(df['Marital_Status'] == 'Widow', 4,
                                                                        np.where(df['Marital_Status'] == 'Alone', 0,
                                                                                 np.where(df['Marital_Status'] == 'Absurd', 0,
                                                                                          np.where(df['Marital_Status'] == 'YOLO', 0, 1))))))))
    return df
df = marital_clean(df,'Marital_Status')                                                                                          
                                                                                        
                                                               
                                                               
                                    
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
```

```python
X = df.drop(['Complain'], axis =1)
y = df['Complain']
plt.scatter(X['Income'],X['NumDealsPurchases'])

plt.xlabel('Income')
plt.ylabel('Number of Deals used')
plt.show()
```

```python
inertias = []
distortions = []

for k in range(1,10):
    model = KMeans(n_clusters=k).fit(X)
    distance = cdist(X, model.cluster_centers_,'euclidean')
    distortion = sum(np.min(distance,axis =1))/X.shape[0]
    distortions.append(distortion)
    inertia = model.inertia_
    inertias.append(inertia)
```

```python
plt.plot(range(1,10),distortions)
plt.title('Distortion Plot')
```

```python
plt.plot(range(1,10),inertias)
plt.title('Inertia Plot')
```

I started the K means by just dropping the complains column in order to troubleshoot the code until it ended up working. I then plotted the inertia and distortion plots both showing an optimal cluster size of around 2 or 3. I then decided to add more columns into the k mean operation in order to cluster through other customer segments that I thought would impact the amount of good purchased. This worked, but the elbow plots are now less similar and almost contradict each other. Based off of these new plots I would say an optimal amount of clusters would be 3 based on the inertia plot, and 2 based on the distortion one. Due to the difficulty of the k means when I added all of the customer variables I decided to cluster buy complaints.  For the k means I plotted the x value as Income and the y as the number of deals used. This shows how the use of coupons vary based on the income of the customer and how this affects the complaint rate. This clearly shows that the majority of the complaints come from customers under or right arounf 100k salary. The more deals used by customers in this range tends to lower the complaint rate. Based on this a good marketing strategy would be to target special deals to customers with lower incomes to raise overall customer satisfaction.
