### Q3

```python
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler


a = cleaned_df.drop('label', axis=1)
b = cleaned_df['label']

a_train, a_test, b_train, b_test = train_test_split(a,b,
                                                    test_size=0.3, random_state=50)
sc = StandardScaler()
a_train = sc.fit_transform(a_train)
a_test = sc.fit_transform(a_test)
```

```python
model = tree.DecisionTreeClassifier(max_depth = 5, random_state=50)
```

```python
model = model.fit(a_train, b_train)
b_pred = model.predict(a_test)
```

```python
print(classification_report(b_test, b_pred))
```

```python
import matplotlib.pyplot as plt
tree.plot_tree(model, feature_names = list(a.columns),
               filled=True, fontsize=8, rounded=True, max_depth = 20)

plt.savefig("tree.svg", format = 'svg', dpi = 100)
plt.figure(figsize=(200, 300))
plt.show()
```
