```python
#Favorite Food
food_chart = df.groupby('Favorite Food').count()
food_chart
```

```python
colors = ['purple','green','blue','orange','pink','yellow','red']
food_chart.plot(kind='pie',y='Timestamp',autopct = '%1.0f%%', colors =colors)
```

This pie chart shows that more students' favorite cuisine is Italian with other cusines or food that falls into multiple cuisines being the runner up.

```python
#Majors
Majors_chart = df.groupby('Major').count()
Majors_chart
```
